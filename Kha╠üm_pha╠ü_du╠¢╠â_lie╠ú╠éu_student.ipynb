````python
# Bước 3.3.1: Phân bố các biến phân loại (Categorical Variables)
# Step 3.3.1: Distribution of Categorical Variables
print("PHÂN TÍCH PHÂN BỐ CÁC BIẾN PHÂN LOẠI")
print("="*50)

# ✅ Tạo subplot cho 4 biến categorical
fig, axes = plt.subplots(2, 2, figsize=(15, 10))   # 4 subplots (2 hàng × 2 cột)
fig.suptitle("Phân bố các biến phân loại", fontsize=16, fontweight='bold')

# ✅ Danh sách các biến cần vẽ và màu sắc
categorical_cols = ["Sex", "BP", "Cholesterol", "Drug"]
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold']

# ✅ Vẽ từng biểu đồ
for i, col in enumerate(categorical_cols):
    row, col_idx = i // 2, i % 2
    ax = axes[row, col_idx]

    # Đếm tần suất
    value_counts = my_data[col].value_counts()

    # Biểu đồ cột
    sns.countplot(data=my_data, x=col, ax=ax, color=colors[i], order=value_counts.index)

    # Tiêu đề và nhãn
    ax.set_title(f'Phân bố {col}', fontweight='bold', fontsize=12)
    ax.set_xlabel(col, fontweight='bold')
    ax.set_ylabel('Số lượng', fontweight='bold')

    # Ghi nhãn số trên cột
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom', fontweight='bold')

    # In thống kê tóm tắt ra console
    print(f"\n{col.upper()}:")
    for category, count in value_counts.items():
        percentage = (count / len(my_data)) * 100
        print(f"  • {category}: {count} bệnh nhân ({percentage:.1f}%)")

# ✅ Căn chỉnh layout và hiển thị
plt.tight_layout()
plt.show()

# ✅ Tổng kết insights
print(f"\nKEY INSIGHTS:")

# Giới tính
sex_counts = my_data['Sex'].value_counts()
if abs(sex_counts.get('M', 0) - sex_counts.get('F', 0)) / len(my_data) < 0.1:
    print("• Giới tính: Phân bố khá cân bằng giữa nam và nữ.")
else:
    print("• Giới tính: Không cân bằng, có sự chênh lệch đáng kể giữa nam và nữ.")

# Huyết áp
bp_mode = my_data['BP'].mode()[0]
print(f"• Huyết áp: Nhóm phổ biến nhất là '{bp_mode}'.")

# Cholesterol
chol_mode = my_data['Cholesterol'].mode()[0]
print(f"• Cholesterol: Nhóm phổ biến nhất là '{chol_mode}'.")

# Drug
drug_mode = my_data['Drug'].mode()[0]
print(f"• Drug distribution: Loại thuốc được kê nhiều nhất là '{drug_mode}'.")
````

````python
print("BÀI 1: Đếm số lượng bệnh nhân theo từng loại thuốc")
print("="*60)

drug_counts = my_data["Drug"].value_counts()   # Đếm số bệnh nhân từng loại thuốc
print(drug_counts)

most_common_drug = drug_counts.idxmax()        # Thuốc có nhiều bệnh nhân nhất
most_common_count = drug_counts.max()

print(f"\n→ Thuốc có nhiều bệnh nhân nhất là: {most_common_drug} ({most_common_count} bệnh nhân)")
````

````python
print("\nBÀI 3: Loại thuốc phổ biến nhất trong từng nhóm huyết áp")
print("="*60)

bp_drug_mode = my_data.groupby("BP")["Drug"].agg(lambda x: x.value_counts().idxmax())
print(bp_drug_mode)

for bp, drug in bp_drug_mode.items():
    print(f"→ Nhóm huyết áp {bp}: thuốc phổ biến nhất là {drug}")
````

````python
print("\nBÀI 4: Bảng chéo giữa giới tính và loại thuốc (tỷ lệ %)")
print("="*60)

crosstab = pd.crosstab(my_data["Sex"], my_data["Drug"], normalize="index") * 100
print(crosstab.round(1))

# Nhận xét nhanh:
print("\nNHẬN XÉT:")
for sex in crosstab.index:
    top_drug = crosstab.loc[sex].idxmax()
    pct = crosstab.loc[sex].max()
    print(f"→ Giới tính {sex}: thuốc phổ biến nhất là {top_drug} ({pct:.1f}%)")
````
