**Mục tiêu nghiên cứu / Research Goal**: Xây dựng một mô hình dự đoán để tìm ra loại thuốc phù hợp nhất cho bệnh nhân mới dựa trên các đặc điểm sinh học và y khoa của họ.

#### Mô tả các Trường Dữ liệu / Data Dictionary

Bộ dữ liệu `drug200.csv` bao gồm **6 trường thông tin** quan trọng:  
The dataset `drug200.csv` contains **6 important fields**:

| Tên trường / Field Name | Kiểu dữ liệu / Data Type | Mô tả / Description | Giá trị có thể / Possible Values |

| **Age** | Số (int) / Numeric | Tuổi của bệnh nhân / Patient's age | 15-74 tuổi / 15-74 years |

| **Drug** | Phân loại / Categorical | Loại thuốc hiệu quả (biến mục tiêu / target variable) | drugA, drugB, drugC, drugX, drugY |

#### Ý nghĩa của từng Biến / Meaning of Each Variable

- **BP**: Huyết áp là chỉ số quan trọng; một số thuốc có thể tác động đến huyết áp / Blood pressure is an important indicator; some drugs can affect BP.  
- **Cholesterol**: Mức cholesterol cao có thể yêu cầu loại thuốc đặc biệt / High cholesterol may require specific drugs.  
- **Na_to_K**: Tỷ lệ này phản ánh cân bằng điện giải trong cơ thể, rất quan trọng trong y khoa / This ratio reflects electrolyte balance, crucial in medicine.  
- **Drug**: Đây là **biến mục tiêu** (target variable) mà chúng ta muốn dự đoán / The **target variable** we aim to predict.

Import các thư viện cần thiết và thiết lập cấu hình để tạo ra các biểu đồ đẹp và chuyên nghiệp.

### Bước 4: Phân tích Quan hệ giữa Biến độc lập và Biến mục tiêu (Bivariate Analysis)

**Mục đích / Purpose**: Khám phá mối quan hệ giữa từng đặc trưng (features) với biến mục tiêu **Drug** để hiểu yếu tố nào ảnh hưởng đến việc lựa chọn thuốc.  
Explore the relationship between each feature and the target variable **Drug** to understand which factors influence drug selection.

#### 4.1. Phân tích Na_to_K theo từng loại thuốc bằng Boxplot / Analyzing Na_to_K by Drug using Boxplot


````python
plt.figure(figsize=(10,6))
sns.boxplot(data=my_data,
            x="Drug",
            y="Na_to_K",
            hue="Drug",
            palette="Set2",
            legend=False)
plt.title("So sánh Tỷ lệ Na/K theo từng loại thuốc", fontsize=14, fontweight='bold')
plt.xlabel("Loại thuốc", fontsize=12)
plt.ylabel("Tỷ lệ Na/K", fontsize=12)
plt.xticks(rotation=45)
plt.show()
````

#### Phân tích Kết quả Boxplot / Boxplot Analysis Results:

**Quan sát quan trọng từ boxplot Na_to_K theo Drug / Key observations from Na_to_K boxplot by Drug:**

- **drugY**: Có **tỷ lệ Na/K cao nhất** (khoảng 25-35), với median ở mức cao  
  **drugY**: Has the **highest Na/K ratio** (around 25-35), with a high median
- **drugX**: Có **tỷ lệ Na/K trung bình** (khoảng 15-25)  
  **drugX**: Has a **medium Na/K ratio** (around 15-25)
- **drugA, drugB, drugC**: Có **tỷ lệ Na/K thấp** (dưới 15), khá tương đồng nhau  
  **drugA, drugB, drugC**: Have **low Na/K ratios** (below 15), fairly similar

**Insight quan trọng / Key insights**:
- Tỷ lệ **Na/K là yếu tố quyết định** trong việc lựa chọn thuốc  
  **Na/K ratio is a determining factor** in drug selection
- Bệnh nhân có **Na/K cao** → chủ yếu được kê **drugY**  
  Patients with **high Na/K** → mostly prescribed **drugY**
- Bệnh nhân có **Na/K thấp** → được kê **drugA/B/C**  
  Patients with **low Na/K** → prescribed **drugA/B/C**
- Đây có thể là **biomarker** quan trọng để dự đoán loại thuốc phù hợp  
  This could be an important **biomarker** for predicting the suitable drug

#### 5.2. Phân tích Age theo từng loại thuốc bằng Violin Plot / Analyzing Age by Drug using Violin Plot


````python
plt.figure(figsize=(10,6))
sns.violinplot(data=my_data,
               x="Drug",
               y="Age",
               hue="Drug",
               palette="viridis",
               inner="quartile",
               legend=False)
plt.title("Phân bố Tuổi theo từng loại thuốc", fontsize=14, fontweight='bold')
plt.xlabel("Loại thuốc", fontsize=12)
plt.ylabel("Tuổi", fontsize=12)
plt.show()
````

#### Phân tích Kết quả Violin Plot / Violin Plot Analysis Results:

**Violin plot kết hợp ưu điểm của boxplot và histogram / Violin plots combine the advantages of boxplots and histograms:**

**Nhận xét về phân bố tuổi theo thuốc / Observations on Age distribution by Drug:**

- **drugY**: Phân bố tuổi **khá rộng** (15-70), không có pattern rõ ràng về tuổi  
  **drugY**: Age distribution is **quite wide** (15-70), no clear age pattern
- **drugX**: Cũng có **phân bố rộng** về tuổi, tương tự drugY  
  **drugX**: Also has a **wide age distribution**, similar to drugY
- **drugA, drugB, drugC**: Có **phân bố hẹp hơn** và tập trung ở các độ tuổi nhất định  
  **drugA, drugB, drugC**: Have a **narrower distribution**, concentrated at certain ages

**So sánh với Boxplot / Comparison with Boxplot**:
- **Violin plot** cho thấy **hình dạng phân bố** (distribution shape) chi tiết hơn  
  **Violin plot** shows the **detailed shape of the distribution**
- **Inner quartiles** hiển thị Q1, median, Q3 như boxplot  
  **Inner quartiles** display Q1, median, Q3 like a boxplot
- **Độ rộng** của violin thể hiện **mật độ** dữ liệu tại từng mức tuổi  
  **Width** of the violin indicates the **density** of data at each age level

**Kết luận / Conclusion**: Tuổi có vẻ **ít ảnh hưởng** đến việc lựa chọn thuốc hơn so với tỷ lệ Na/K  
**Age seems to have less influence** on drug selection compared to the Na/K ratio

#### 5.3. Ma trận Tương quan giữa BP, Cholesterol và Drug / Correlation Matrix between BP, Cholesterol, and Drug


````python
crosstab = pd.crosstab(index=[my_data["BP"], my_data["Cholesterol"]],
                       columns=my_data["Drug"])
plt.figure(figsize=(10,8))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu",
            cbar_kws={'label': 'Số lượng bệnh nhân'})
plt.title("Ma trận Tương quan: Tần suất kê thuốc theo Huyết áp & Cholesterol", fontsize=14, fontweight='bold')
plt.xlabel("Loại thuốc", fontsize=12)
plt.ylabel("Huyết áp & Cholesterol", fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
````

#### Phân tích Kết quả Heatmap (Heatmap Analysis Results)

**Heatmap cho thấy patterns rất thú vị (The heatmap reveals very interesting patterns):**

**Quan sát chính (Key Observations):**
1. **drugY**: Được kê chủ yếu cho bệnh nhân có **Low BP** (huyết áp thấp)  
   **drugY**: Mainly prescribed to patients with **Low BP**
2. **drugX**: Được kê cho bệnh nhân có **Normal BP**  
   **drugX**: Prescribed to patients with **Normal BP**
3. **drugA, drugB, drugC**: Được kê chủ yếu cho bệnh nhân có **High BP** (huyết áp cao)  
   **drugA, drugB, drugC**: Mainly prescribed to patients with **High BP**

**Pattern quan trọng về Cholesterol (Important Pattern regarding Cholesterol):**

#### 7.2. Stacked Bar Chart: Tỷ lệ thuốc theo giới tính / Stacked Bar Chart: Drug distribution by Sex


````python
sex_drug_prop = pd.crosstab(my_data["Sex"],
                            my_data["Drug"],
                            normalize="index")

plt.figure(figsize=(10,6))
sex_drug_prop.plot(kind="bar",
                   stacked=True,
                   cmap="tab10",
                   figsize=(10,6))
plt.title("Drug Proportion by Sex / Tỷ lệ Loại thuốc theo Giới tính", fontsize=14, fontweight='bold')
plt.xlabel("Giới tính", fontsize=12)
plt.ylabel("Tỷ lệ (%)", fontsize=12)
plt.legend(title="Loại thuốc", bbox_to_anchor=(1.05,1), loc='upper left')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
print("Bảng tỷ lệ phần trăm thuốc theo giới tính:")
print(sex_drug_prop.round(3))
````

````python
from sklearn.preprocessing import LabelEncoder
from mpl_toolkits.mplot3d import Axes3D

encoder = LabelEncoder()
my_data['BP_code'] = encoder.fit_transform(my_data['BP'])

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(my_data['Age'], my_data['Na_to_K'], my_data['BP_code'],
                c=pd.Categorical(my_data['Drug']).codes, cmap='Set1', s=60)
ax.set_xlabel("Age")
ax.set_ylabel("Na_to_K")
ax.set_zlabel("BP_code")
plt.title("3D Scatterplot: Age - Na_to_K - BP_code (Color by Drug)", fontsize=13)
plt.legend(*sc.legend_elements(), title="Drug")
plt.show()
````

````python
ct = pd.crosstab([my_data["Sex"], my_data["BP"], my_data["Cholesterol"]], my_data["Drug"])
plt.figure(figsize=(12,8))
sns.heatmap(ct, annot=True, fmt="d", cmap="YlOrRd", cbar_kws={'label': 'Số lượng'})
plt.title("Heatmap: Tần suất kê thuốc theo (Sex, BP, Cholesterol)", fontsize=14, fontweight='bold')
plt.xlabel("Loại thuốc")
plt.ylabel("(Sex, BP, Cholesterol)")
plt.tight_layout()
plt.show()
````
